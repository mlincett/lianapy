```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
low, high = 0, 10
```

```python
# Random sample of integers
sample = np.random.randint(low, 1+high, size=100)

# Trivial flat distribution of integers
sample_flat = np.arange(low, 1 + high)
```

```python
def discrete_histogram(sample):
    lo, hi = np.min(sample), np.max(sample)
    r = (lo -0.5, hi +0.5)
    b = 1 + hi - lo
    return { 'range': r, 'bins': b }
```

```python
fig, ax = plt.subplots(1,2, figsize=(8,4))
H0 = ax[0].hist(sample_flat, **discrete_histogram(sample_flat))
H1 = ax[1].hist(sample     , **discrete_histogram(sample)     )
```
